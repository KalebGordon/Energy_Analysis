```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("capacity_factor.csv")
df
```

```python
# Capacity Factor (%)
plt.figure(figsize=(10,6))
sns.barplot(x='Energy Type', y='CF (%)', data=df, palette='viridis')
plt.xlabel('Energy Type')
plt.ylabel('Capacity Factor (%)')
plt.title("Energy Source Comparison: Capacity Factor")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# GHG Emission Rate (g/kWh)
plt.figure(figsize=(10,6))
sns.barplot(x='Energy Type', y='GHG Emission Rate (g/kWh)', data=df, palette='viridis')
plt.xlabel('Energy Type')
plt.ylabel('GHGeq Emission Rate (g/kWh)')
plt.title("Energy Source Comparison: Capacity Factor")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Waste (Metric Tons)
plt.figure(figsize=(10,6))
sns.barplot(x='Energy Type', y='Physical Waste Per Year in the US(Metric Tons)', data=df, palette='viridis')
plt.xlabel('Energy Type')
plt.ylabel('GHGeq Emission Rate (g/kWh)')
plt.title("Energy Source Comparison: Capacity Factor")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
